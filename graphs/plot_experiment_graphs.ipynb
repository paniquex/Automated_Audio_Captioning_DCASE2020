```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
without_mixup = pd.read_csv('./data/run-Mar31_14-39-41_pnqwithout_mixup_without_apex-tag-spider.csv')
without_mixup2 = pd.read_csv('./data/run-Mar31_20-05-22_pnqwithout_mixup_without_apex-tag-spider.csv')
without_mixup = pd.concat([without_mixup, without_mixup2]).reset_index()
with_mixup = pd.read_csv('./data/run-Apr01_01-37-43_pnqwith_mixup_without_apex-tag-spider.csv')
```

```python
without_mixup.loc[100:, 'Step'] += 100
```

```python
np.max(without_mixup['Value']), np.max(with_mixup['Value'])
```

```python
sns.lineplot(x='Step', y='Value', data=without_mixup)
sns.lineplot(x='Step', y='Value', data=with_mixup)

plt.legend(['without_mixup', 'with_mixup'])
plt.title('Spider metric')
```
